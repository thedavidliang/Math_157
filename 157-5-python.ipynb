```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cmath
```

Problem 5 - Part 1

```python
# converts (x,y) into complex number

def complexify(x, y, xmax, ymax, amin, amax, bmin, bmax):
    xfrac = float(x) / float(xmax)
    yfrac = float(y) / float(ymax)
    a = amin + xfrac * (amax - amin)
    b = bmin + yfrac * (bmax - bmin)
    return a + b*1j
```

```python
complexify(1,1,3,2,0,10,0,3)
```

Problem 5 - Part 2

```python
# let z_n = f(z_{n-1})
# returns smallest such that |z_n| > T, 
#   or returns nmax if z_n converges to zero

def compute_value(z, f, T, nmax):
    if nmax == 0:
        return 0  
    elif (abs(z) > T):
        return 0
    else:
        return 1 + compute_value(f(z), f, T, (nmax - 1))
```

Problem 5 - Part 3

```python
# draws fractal according to values from compute_value

def draw_fractal(f, filename, xmax, ymax, amin, amax, bmin, bmax, T, nmax):
    f_new = lambda x,y: compute_value(complexify(x,y,xmax,ymax,amin,amax,bmin,bmax), 
                                      f, T, nmax)
    fractal_plot = np.fromfunction(np.vectorize(f_new), (xmax,ymax))
    mpimg.imsave(filename, fractal_plot)

    plt.imshow(fractal_plot)
    plt.show()
```
